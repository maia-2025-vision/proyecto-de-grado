```python
name_to_mnemo = {
    "frcn_resnet101_full_imgs": "FRC-101",
    "frcnn-resnet101-full-imgs": "FRC-101",
    "frcnn-resnet50-full-imgs": "FRC-50",
    "frcn_resnet50_full_imgs": "FRC-50",
    "herdnet_v1_fullimgs": "HN-1",
    "herdnet_v1_full_imgs": "HN-1",
    "herdnet_v2_full_imgs": "HN-2",
    "herdnet_v2_hn2": "HN-2-hn",
    "herdnet_v2_hn2_full_imgs": "HN-2-hn",
    "herdnet_v3_full_imgs": "HN-3",
    "herdnet_v4_full_imgs": "HN-4",
}
```

```python
parts = []
for model_dir in sorted(BASE_PATH.glob("*")):
    dir_name = model_dir.name
    if (
        dir_name.endswith("-xlsx")
        or dir_name.startswith(".")
        or "cnts_metrics" in dir_name
        or dir_name == "frcnn-resnet50"
        or dir_name in ("herdnet_v0", "frcnn-resnet50", "herdnet_v2_hn", "herdnet_v1", "herdnet_v2")
    ):
        continue
    # print(list((BASE_PATH / dir_name).glob('*')))
    part = pd.read_csv(BASE_PATH / dir_name / "metrics_results.csv")
    part["Modelo"] = name_to_mnemo[dir_name]
    parts.append(part[part["species"] == "binary"])

df = pd.concat(parts)

for metric in ["recall", "precision", "f1_score", "ap"]:
    df[metric] = np.round(df[metric] * 100, 1)
```

```python
df1 = df[["Modelo", "recall", "precision", "f1_score", "ap"]].rename(
    columns={"recall": "Recall", "precision": "Precisión", "ap": "AP", "f1_score": "Puntaje\nF1"}
)
df1
```

```python
reload(lu)
print(
    lu.df_to_latex(df1, caption="Métricas de Recall/Precisión binarias sobre el conjunto de test")
)
```

```python
from pathlib import Path

model_dirs = list(Path("data/cnt_metrics_v2").glob("*"))
model_dirs
```

```python
parts = []


for model_dir in sorted(model_dirs):
    name = model_dir.name
    full_df = pd.read_csv(model_dir / "aggregated_count_metrics_v2.csv")
    part = full_df[full_df["Especie"] == "binaria"].copy()
    part["Modelo"] = name_to_mnemo[name]

    parts.append(part)

df = pd.concat(parts)
df["Σ(Ĉ)"] = df["Σ(Ĉ)"].astype(int)

df = df["Modelo,M(Ĉ),Σ(Ĉ),ME,MAE,RMSE,MPE,MAPE".split(",")]

# Renames to prepare for latex output
df = df.rename(
    columns={
        "M(Ĉ)": r"$\mbox{M}(\hat{C})$",
        "Σ(Ĉ)": r"$\mathbf{\Sigma}(\hat{C})$",
        "MPE": "MPE\n[\\%]",
        "MAPE": "MAPE\n[\\%]",
    }
)
df
```

```python
{
    name: name.replace("_full_imgs", "").replace("herdnet_v", "HN-").replace("frcn_resnet", "FRC-")
    for name in df["Modelo"].values
}
```

```python
df.columns = [col.replace("Σ", r"$\Sigma$") for col in df.columns]
print(
    lu.df_to_latex(
        df,
        caption=(
            "Métricas de Conteo, caso binario, para los modelos entrenados. "
            "El promedio real del conteo de animales en el conjunto de test es "
            r"$\mbox{M}(C^*) = $8.9 y la suma de los conteos es $\Sigma(C^*) = 2999$"
        ),
    )
)
```

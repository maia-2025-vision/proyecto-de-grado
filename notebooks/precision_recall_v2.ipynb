```python
import os
import pandas as pd
from pathlib import Path


if Path.cwd().name == "notebooks":
    os.chdir("..")
```

```python
detections = pd.read_csv("data/test_results/herdnet_v2_hn2/detections.csv")
display(detections)

print("unique images:", detections["images"].nunique())
print(detections["labels"].isna().sum())
```

```python
gt_boxes = pd.read_csv("data/groundtruth/csv/test_big_size_A_B_E_K_WH_WB-fixed-header.csv")
gt_boxes["gt_x"] = (gt_boxes["x_min"] + gt_boxes["x_max"]) * 0.5
gt_boxes["gt_y"] = (gt_boxes["y_min"] + gt_boxes["y_max"]) * 0.5
gt_boxes["gt_width"] = gt_boxes["x_max"] - gt_boxes["x_min"]
gt_boxes["gt_height"] = gt_boxes["y_max"] - gt_boxes["y_min"]
display(gt_boxes.head(5))
print("unique images:", gt_boxes["images"].nunique())
```

```python
gt_by_img = gt_boxes["images"].value_counts().reset_index()
detections_by_img = detections["images"].value_counts().reset_index()

merged_cnts = gt_by_img.merge(detections_by_img, on="images")
merged_cnts
```

```python
image = "39220a3bb1a3f74175d10221fce6926bb66969bf.JPG"
gt_df = gt_boxes[gt_boxes["images"] == image].sort_values("gt_x")
gt_df[["images", "gt_x", "gt_y", "labels"]]
```

```python
# !ls -lrt data/
gt_bb_v2 = pd.read_csv("data/gt-preprocessed/csv/test_big_size_A_B_E_K_WH_WB-points.csv")
gt_df_v2 = gt_bb_v2[gt_bb_v2["images"] == image].sort_values("x")
gt_df_v2
```

```python
detect_df = detections[detections["images"] == image]
detect_df = detect_df[~detect_df["labels"].isna()].sort_values("x")
detect_df[["images", "x", "y"]]
```

```python
n_gt = len(gt_df)
n_detects = len(detect_df)
n_gt, n_detects
```

```python
import numpy as np

tol = 0.50  # 50% in

hit_mat = np.zeros(shape=(n_gt, n_detects), dtype=np.bool)
dist_mat = np.zeros(shape=(n_gt, n_detects), dtype=float)
```

```python
for gt_i in range(n_gt):
    gt_row = gt_df.iloc[gt_i]
    hit_x_min = gt_row["gt_x"] - 0.5 * tol * gt_row["gt_width"]
    hit_x_max = gt_row["gt_x"] + 0.5 * tol * gt_row["gt_width"]
    hit_y_min = gt_row["gt_y"] - 0.5 * tol * gt_row["gt_height"]
    hit_y_max = gt_row["gt_y"] + 0.5 * tol * gt_row["gt_height"]

    det_x = detect_df["x"].values
    det_y = detect_df["y"].values

    for d_j in range(n_detects):
        det_x = detect_df["x"].iloc[d_j]
        det_y = detect_df["y"].iloc[d_j]
        hit_mat[gt_i, d_j] = (
            (hit_x_min <= det_x)
            & (det_x <= hit_x_max)
            & (hit_y_min <= det_y)
            & (det_y <= hit_y_max)
        )

    # dist_mat[gt_i, :] = np.sqrt( (gt_row['gt_x'] - det_x)**2 + (gt_row['gt_y'] - det_y) ** 2)
```
